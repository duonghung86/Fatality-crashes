```python
# Basic packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Import dataset

```python
#sheet='Weight'
sheet='BorderlineSMOTE'
df=pd.read_excel('SA of input full.xlsx',sheet_name=sheet)
print(df.shape)
df.head()
```

```python
df=df.iloc[:,1:].copy()
df.head()
```

```python
df.iloc[:,2].name[:-2]
```

```python
# filter the strong correlation factor
key_facs=pd.DataFrame([[0,0,0]],columns=['Name','SD','R'])
#print(key_facs)
for i in range(150):
    cor_coe=(np.corrcoef(df.iloc[:,2*i], df.iloc[:,2*i+1])[0,1])
    sd=df.iloc[:,2*i+1].std()
    if cor_coe**2>0.8:
        key_facs=pd.concat([key_facs,pd.DataFrame([[df.iloc[:,2*i].name[:-2],sd,cor_coe]],columns=['Name','SD','R'])])

key_facs=key_facs.iloc[1:,:].copy()        
print(key_facs)
```

```python
key_facs.sort_values(by='SD',ascending=False).head(10)
```

```python
top10=key_facs.sort_values(by='SD',ascending=False).head(10)
top10
```

```python
top10['Cor_typ']=np.where(top10.R>0,'Positive Correlation','Negative Correlation')
top10
```

```python
neg=top10[top10.R<0].Name.values
neg
```

```python
top10f=[x+'_Y' for x in top10.Name]
top10f
```

```python
df_y=df[top10f]
df_y
```

```python
melted_df_y=pd.melt(df_y)
melted_df_y.head()
```

```python
melted_df_y.variable=[x[:-2] for x in melted_df_y.variable]
melted_df_y.head()
```

```python
melted_df_y['Correlation']=np.where(melted_df_y.variable.isin(neg),'Negative','Positive')
melted_df_y
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=melted_df_y, orient="h", y='variable', x= 'value',hue='Correlation')
plt.xlabel('Probability of Fatality')
plt.title('Top 10 risk factors for '+sheet+' model')
plt.xticks(np.arange(-0.0,1.1,0.2))
plt.savefig('Top10 for '+sheet+'.png',dpi=150,bbox_inches='tight')
plt.show()
```
