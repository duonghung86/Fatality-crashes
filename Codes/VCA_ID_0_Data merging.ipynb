# Import the person dataset

```python
df_person=pd.read_csv(filelist[6])
df_person.head()
```

```python
df_person.shape
```

```python
# Percentage of missing values for each feature
mvr=df_person.isna().sum()/len(df_person.Crash_ID)
valid_feat=mvr[mvr<0.2].index
print(len(valid_feat))
valid_feat
```

```python
keeps=['Crash_ID', 'Unit_Nbr', 'Prsn_Nbr', 'Prsn_Type_ID',
       'Prsn_Occpnt_Pos_ID', 
       'Prsn_Injry_Sev_ID', 'Prsn_Age', 'Prsn_Ethnicity_ID', 'Prsn_Gndr_ID',
       'Prsn_Ejct_ID', 'Prsn_Rest_ID', 'Prsn_Airbag_ID', 'Prsn_Helmet_ID',
       'Prsn_Drg_Rslt_ID']
df_person=df_person[keeps].copy()
df_person.head()
```

```python
df_final=df_person.copy()
```
