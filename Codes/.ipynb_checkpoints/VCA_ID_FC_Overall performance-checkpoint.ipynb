```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('Overall performance lastest.xlsx')
df.head()
```

```python
df['TNR']=df.tn/(df.tn+df.fp)
df['BA']=(df.TNR+df.recall)/2
df['F1 score']=2*df.recall*df.precision/(df.recall+df.precision)
df
```

```python
df=df[4:]
```

```python
df=df[['accuracy','precision','recall','auc','BA','F1 score']]
df
```

```python
df[['accuracy','precision','recall','auc','BA']]=df[['accuracy','precision','recall','auc','BA']]*100
```

```python
df=df.round(3)
```

```python
df_train=df[0::2]
df_train
```

```python
models=['Optimized weight', 'Random oversampling', 'SMOTE', 'BorderlineSMOTE','ADASYN']
```

```python
df_train.index=models
df_train
```

```python
df_test=df[1::2]
df_test.index=models
df_test
```

```python
df_delta=df_train-df_test
df_delta
```
