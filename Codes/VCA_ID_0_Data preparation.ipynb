```python
from io import StringIO
from io import BytesIO
from zipfile import ZipFile
import urllib.request
```

## Function to obtain data set from zip link

```python
def zip2df(link):
    # Open the url link
    url = urllib.request.urlopen(link)
    
    # Open as zip file
    with ZipFile(BytesIO(url.read())) as my_zip_file:
        for contained_file in my_zip_file.namelist():
            fzip=my_zip_file.open(contained_file)
            print(contained_file)
            data=fzip.read()
    # Convert bytes object to string object
    s=str(data,'utf-8')
    data = StringIO(s) 
    # convert it to pandas DataFrame as normal csv file
    df=pd.read_csv(data)
    return df
```

# Mapping the data

## Get the list of mapping sheets

```python
term_link='https://github.com/duonghung86/Fatality-crashes/raw/master/Data/Term.xlsx'
xls = pd.ExcelFile(term_link) # help reduce time when read multiple sheets
sheets=xls.sheet_names
```

```python
# Link between feature name and its coding sheet

link_c=pd.read_excel(xls,sheet_name='Crash File Specification',header=6,usecols=['Column Header','Description','Data Values with Valid Years if Applicable'])
link_c.head()
```

```python
# Link between feature name and its coding sheet
link_p=pd.read_excel(xls,sheet_name='PrimaryPerson File Spec.',header=5,usecols=['Column Header','Description','Data Values with Valid Years if Applicable'])
link_p.head()
```

```python
# Link between feature name and its coding sheet
link_u=pd.read_excel(xls,sheet_name='Unit File Specification',header=6,usecols=['Column Header','Description','Data Values with Valid Years if Applicable'])
link_u.head()
```

```python
feats=df_final.columns
```

```python
feats.shape
```

```python
# Combine all link
link_sheet=pd.concat([link_p,link_c,link_u]).reset_index(drop=True)
link_sheet=link_sheet[link_sheet['Column Header'].isin(feats)].reset_index(drop=True)
link_sheet
```

```python
link_sheet.to_csv('Chosen features.csv',index=False)
```

```python
# Drop na features
link_sheet.dropna(inplace=True)
# Drop the yes-no choice
link_sheet=link_sheet[link_sheet['Data Values with Valid Years if Applicable']!='YES_NO_CHOICE_LKP']
link_sheet.reset_index(drop=True, inplace=True)
link_sheet
```

## Map the categorical features

```python
#Features need mapping
for i in link_sheet.index:
    print(i/len(link_sheet)*100)
    feat=link_sheet['Column Header'][i]
    sfeat=link_sheet.iloc[i,2]
    print(feat)
    df_utype=pd.read_excel(xls,sheet_name=sfeat)
    df_utype.set_index(df_utype.columns[0],inplace=True)
    df_utype=df_utype.iloc[:,0]
    #remove duplicated index
    df_utype=df_utype[~df_utype.index.duplicated(keep='first')]
    #print(df_utype)
    df_final[feat]=df_final[feat].map(df_utype)
df_final.head()
```
