# Mapping

## Get the link

```python
# Link between feature name and its coding sheet
link_c=pd.read_excel("Term.xlsx",sheet_name='Crash File Specification',header=6,usecols=['Column Header','Description','Data Values with Valid Years if Applicable'])
link_c.head()
```

```python
# Link between feature name and its coding sheet
link_p=pd.read_excel("Term.xlsx",sheet_name='PrimaryPerson File Spec.',header=5,usecols=['Column Header','Description','Data Values with Valid Years if Applicable'])
link_p.head()
```

```python
# Link between feature name and its coding sheet
link_u=pd.read_excel("Term.xlsx",sheet_name='Unit File Specification',header=6,usecols=['Column Header','Description','Data Values with Valid Years if Applicable'])
link_u.head()
```

```python
# Combine all link
link_sheet=pd.concat([link_p,link_c,link_u]).reset_index(drop=True)
link_sheet=link_sheet[link_sheet['Column Header'].isin(feats)].reset_index(drop=True)
link_sheet
```

```python
link_sheet=link_sheet.drop([14,33,34]).reset_index(drop=True)
link_sheet
```

```python
link_sheet.to_csv('Chosen features.csv')
```

## Link filtering

```python
# Drop na features
link_sheet.dropna(inplace=True)
# Drop the yes-no choice
link_sheet=link_sheet[link_sheet['Data Values with Valid Years if Applicable']!='YES_NO_CHOICE_LKP']
link_sheet.reset_index(drop=True, inplace=True)
link_sheet
```

## Mapping the ID

```python
#Features need mapping
for i in link_sheet.index:
    print(i/len(link_sheet)*100)
    feat=link_sheet['Column Header'][i]
    sfeat=link_sheet.iloc[i,2]
    print(feat)
    df_utype=pd.read_excel("Term.xlsx",sheet_name=sfeat)
    df_utype.set_index(df_utype.columns[0],inplace=True)
    df_utype=df_utype.iloc[:,0]
    #remove duplicated index
    df_utype=df_utype[~df_utype.index.duplicated(keep='first')]
    #print(df_utype)
    df[feat]=df[feat].map(df_utype)
df.head()
```

```python
df.to_csv('Mapped combined data.csv',index=False)
```
